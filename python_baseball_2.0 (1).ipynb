Import files / necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
batting = pd.read_csv("C:/Users/pgiom/Downloads/R/Batting.csv")
pitching = pd.read_csv("C:/Users/pgiom/Downloads/Pitching.csv")
salaries = pd.read_csv("C:/Users/pgiom/Downloads/Salaries.csv")
people = pd.read_csv("C:/Users/pgiom/Downloads/R/People_CSV.csv")
```

Merging files 

```python
merge = pd.merge(people, batting, on = 'playerID', how = 'left')
merge.head(5)
```

```python
merge_2 = pd.merge(merge, pitching, on = 'playerID')
merge_2.head(5)
```

```python
final = pd.merge(merge_2, salaries, on = 'playerID')
final.head(5)
```

```python
final.info()
```

```python
final.isna().sum()
```

replace missing values with median since it is less affected by outliers

```python
# only include the numeric columns for median computation
numeric_cols = final.select_dtypes(include=[np.number])

# get median only for the numeric columns
numeric_medians = numeric_cols.median()

# Fill NA values w/ median only in numeric columns 
final[numeric_cols.columns] = numeric_cols.fillna(numeric_medians)
```

```python
final.describe().T
```

The standard deviations above emphasize a strong correlation of variables. Using scatterplots I will show the strongest correlations to each other.

```python
sns.scatterplot(data=final, x="height", y="salary")
plt.title('Heights of MLB Players with the Highest Salaries')
plt.xlabel('Height (inches)')
plt.ylabel('Salary ($)')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

```

#another height graph, this one with boxplots.

```python
# Convert 'salary' to numeric
final['salary'] = pd.to_numeric(final['salary'], errors='coerce')

# Define tall and short players based on height percentile
height_75th_percentile = final['height'].quantile(0.75)
height_25th_percentile = final['height'].quantile(0.25)
tall_players = final[final['height'] >= height_75th_percentile]
short_players = final[final['height'] <= height_25th_percentile]

# Define the salary ranges for both tall and short players
low_salary_tall = tall_players[(tall_players['salary'] >= 0) & (tall_players['salary'] <= 5e6)]
mid_salary_tall = tall_players[(tall_players['salary'] > 5e6) & (tall_players['salary'] < 1e7)]
high_salary_tall = tall_players[(tall_players['salary'] >= 1e7)]

low_salary_short = short_players[(short_players['salary'] >= 0) & (short_players['salary'] <= 5e6)]
mid_salary_short = short_players[(short_players['salary'] > 5e6) & (short_players['salary'] < 1e7)]
high_salary_short = short_players[(short_players['salary'] >= 1e7)]

# Plot settings
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 18), sharex='col', sharey='row')

# Salary range titles
salary_titles = ['0 to 5 Million', '>5 Million to <10 Million', '>=10 Million']

# Plot the boxplots for tall players
for i, (ax, title) in enumerate(zip(axes[:, 0], salary_titles)):
    salary_data = [low_salary_tall, mid_salary_tall, high_salary_tall][i]['salary'].dropna()
    ax.boxplot(salary_data, vert=True, patch_artist=True)
    ax.set_title(f'Tall Players Salaries {title}')
    ax.set_ylabel('Salary')

# Plot the boxplots for short players
for i, (ax, title) in enumerate(zip(axes[:, 1], salary_titles)):
    salary_data = [low_salary_short, mid_salary_short, high_salary_short][i]['salary'].dropna()
    ax.boxplot(salary_data, vert=True, patch_artist=True)
    ax.set_title(f'Short Players Salaries {title}')

# Set y-axis to show the salary in millions for all subplots
for ax in axes.flatten():
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,.1f}M".format(x/1e6)))

plt.tight_layout()
plt.show()

```

Linear Regression Analysis: Comparing specific features below (x) to the (y) "salary"

```python
#let's declare our features (independent variables, x) and the target (dependent variable, y)
x = final[['weight', 'height','G_x','AB','R_x','H_x' ,'X2B','X3B','HR_x','RBI','SB','CS', 'BB_x', 'SO_x', 'IBB_x', 'HBP_x','SH_x','BFP', 'HR_y']]
y = final['salary']
```

```python
# with sklearn
import sklearn
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
import statsmodels.api as sm
x = sm.add_constant(x) # adding a constant
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```

Using sklearn and statsmodel the results were consistent between both. Some of the variables with negative influence on the salary include height, AB, X2B, X3B, RBI, SB, CS, HBP On the contrary, the other variables have a positive relationship with the salaries of players.

Side note: The relationship between G_x and BB_x is not statistically significant since p>0.05
